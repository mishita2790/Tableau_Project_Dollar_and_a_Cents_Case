# Job Selection Data Wrangling

```python
import pandas as pd
from bs4 import BeautifulSoup
import bs4
import requests
```

```python
temp_dict = {}
df = pd.DataFrame()

url = 'http://money.cnn.com/2016/03/23/pf/gender-pay-gap/index.html'

r  = requests.get(url)
soup = BeautifulSoup(open('../PayGap Raw Data/Gender pay gap.html'), 'lxml')

for p in soup.find('p', text=' 1. Computer programmer - 72 cents '):
    temp_text = str(p).split('-')
    temp_dict[0] = temp_text[0].split('.')[1]
    temp_dict[1] = float(temp_text[1].split(' ')[1]) * 0.01
    df = df.append(temp_dict, ignore_index=True)
    for i in range(0, 20):
        p = p.findNext('p')
        if i == 2 or i == 6:
            temp_dict[0] = p.find('a').get_text()
            temp_text = str(p).split(' ')
            temp_dict[1] = float(temp_text[5]) * 0.01
        
        elif i == 9:
            continue
        
        elif i >= 0 and i < 9:
            temp_text = str(p).split('-')
            temp_dict[0] = temp_text[0].split('.')[1]
            temp_dict[1] = float(temp_text[1].split(' ')[1]) * 0.01
            
        else:
            temp_text = str(p).split('-')
            temp_dict[0] = temp_text[0].split('.')[1]
            temp_dict[1] = temp_text[1].split('$')[1]
        
        df = df.append(temp_dict, ignore_index=True)
        df
        
cols = ['Occupation','Women']
df.columns = cols

df
```

```python
df['Women'].replace(regex=True,inplace=True,to_replace=r'</p>',value=r'')
df
```

```python
df['Men'] = 1
df
```

```python
df.to_csv('../Cleaned Data/Job Selection Data.csv')
```
