# Pay Gap By State Data Clean

```python
import pandas as pd
```

```python
states = ['New York','North Dacota','California','Wyoming']

ny = pd.read_excel('../PayGap Raw Data/' + states[0] + '.xls')
nd = pd.read_excel('../PayGap Raw Data/' + states[1] + '.xls')
ca = pd.read_excel('../PayGap Raw Data/' + states[2] + '.xls')
wy = pd.read_excel('../PayGap Raw Data/' + states[3] + '.xls')

wy
    
```

```python
ny = ny.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 4', 'Unnamed: 5','Unnamed: 6'], axis=1)
nd = nd.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 4', 'Unnamed: 5','Unnamed: 6'], axis=1)
ca = ca.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 4', 'Unnamed: 5','Unnamed: 6'], axis=1)
wy = wy.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 4', 'Unnamed: 5','Unnamed: 6'], axis=1)
ca
```

```python
ny = ny.dropna()
nd = nd.dropna()
ca = ca.dropna()
wy = wy.dropna()
nd
```

```python
ny.columns = ['Description','Number']
nd.columns = ['Description','Number']
ca.columns = ['Description','Number']
wy.columns = ['Description','Number']
ny.index = range(len(ny))
nd.index = range(len(nd))
ca.index = range(len(ca))
wy.index = range(len(wy))
ca
```

```python
ny2 = ny[18:]
ny = ny[:18]
nd2 = nd[18:]
nd = nd[:18]
ca2 = ca[18:]
ca = ca[:18]
wy2 = wy[18:]
wy = wy[:18]
wy2
```

```python
wy
```

```python
ny3 = pd.merge(ny, ny2, how='inner', on=['Description'])
ny3
```

```python
nd3 = pd.merge(nd, nd2, how='inner', on=['Description'])
ca3 = pd.merge(ca, ca2, how='inner', on=['Description'])
wy3 = pd.merge(wy, wy2, how='inner', on=['Description'])
```

```python
ny3.columns = ['Description','Male','Female']
nd3.columns = ['Description','Male','Female']
ca3.columns = ['Description','Male','Female']
wy3.columns = ['Description','Male','Female']
```

```python
wy3
```

```python
ny3.to_csv('../Cleaned Data/New York Pay Gap.csv')
```

```python
nd3.to_csv('../Cleaned Data/North Dacota Pay Gap.csv')
ca3.to_csv('../Cleaned Data/California Pay Gap.csv')
wy3.to_csv('../Cleaned Data/Wyoming Pay Gap.csv')
```
