# Prize Moeny Data

```python
import pandas as pd
from bs4 import BeautifulSoup
import bs4
import requests
```

```python
temp_dict = {}
df = pd.DataFrame()

url = 'https://en.wikipedia.org/wiki/List_of_tennis_tournaments'

r  = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib') 
i = 0

for table in soup.find_all('table', attrs={'style':True}):
    for tr in table.find_all('tr'):
        if i == 0:
            temp_dict[0] = 'Male'
        else:
            temp_dict[0] = 'Female'
        j = 1
        for td in tr.find_all('td'):
            if j == 3 or j == 4 or j == 9:
                temp_dict[j] = td.get_text()
            j += 1
        df = df.append(temp_dict, ignore_index=True)
    i += 1

cols = ['Sex','Type','Tournament','Prize_Money']
df.columns = cols

df.head()
```

```python
df
```

```python
df = df.dropna()
df.Prize_Money = df.Prize_Money.str.rstrip().str[1:]
df
```

```python
df.to_csv('../Cleaned Data/Major Tennis Tournament Prize Money Data.csv', encoding='utf-8')
```
